```python
def initialisation (n0,n1,n2):
    """

    Args:
        n0 (_type_): _description_
        n1 (_type_): _description_
        n2 (_type_): _description_

    Returns:
        _type_: _description_
    """
    
    
    W=np.random.randn(X.shape[1], 1)
    b=np.random.randn(1)
    return W, b
W, b = initialisation(X)
```

```python
# calculer la fontion Z
def model (X, W, b):
    z=X.dot(W) + b
    A=1/(1+np.exp(-z))
    return A
```

```python
def log_loss(A,y):
    return -((np.dot(y.T,np.log(A)) + np.dot((1 - y).T, np.log(1 - A)))/ y.shape[0])[0,0]
```

```python
def gradient (X, y, A):
    m = y.shape[0]
    dW = (1/m) * X.T.dot(A - y)
    db = (1/m) * np.sum(A - y)
    return dW, db   
```

```python
def mise_a_jour (W, b, dW, db, learning_rate):
    W = W - learning_rate * dW
    b = b - learning_rate * db
    return (W, b)
```

```python
def prediction(X, W, b):
    A = model(X, W, b)
    return np.where(A >= 0.5, 1, 0)
```

```python
def artificial_neural_network(X, y, learning_rate=0.1, n_iterations=100):
    W, b = initialisation(X)
    Loss=[]
    for i in range(n_iterations):
        A = model(X, W, b)
        loss = log_loss(A, y)
        Loss.append(loss)
        dW, db = gradient(X, y, A)
        W, b = mise_a_jour(W, b, dW, db, learning_rate)
        #if i % 100 == 0:
            #print(f"Iteration {i}, Loss: {loss}")
    #return W, b, Loss 
    y_pred=prediction(X, W, b)
    print(accuracy_score(y, y_pred))
    plt.plot(Loss)
    plt.show()
    return W, b
    
```
